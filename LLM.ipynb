```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import pandas as pd
```

```python
toy_corpus = ["the fat cat sat on the mat","the big cat slept","the dog chased the cat"]
```

```python
vectorizer = TfidfVectorizer()
corpus_tfidf = vectorizer.fit_transform(toy_corpus)
print("vocabulary size >",len(vectorizer.vocabulary_.keys()))
print("the document-tern matrix shape >",corpus_tfidf.shape)
df = pd.DataFrame(np.round(corpus_tfidf.toarray(),2))
df.columns=vectorizer.get_feature_names_out()
print(df)
```

```python
import nltk
from nltk.corpus import gutenberg
from nltk.lm import MLE
from nltk.lm.preprocessing import padded_everygram_pipeline
nltk.download('gutenberg')
nltk.download('punkt')
macbeth = gutenberg.sents('shakespeare-macbeth.txt')
model,vocab = padded_everygram_pipeline(2,macbeth)
lm = MLE(2)
lm.fit(model,vocab)
print(list(lm.vocab)[:10])
print("the number of words > ",len(lm.vocab))
```

```python
print("the freq of word 'macbeth' is ",lm.counts['Macbeth'])
print("the lang model probability score of macbeth is ",lm.score('Macbeth'))
print("the number of times 'Macbeth' follows 'Enter' is ",lm.counts[['Enter']]['Macbeth'])
print("p(Macbeth | Enter) is ",lm.score('Macbeth',['Enter']))
print("p(shaking|for) is",lm.score('shaking',['for']))
```

```python
lm.generate(10,text_seed=['<s>'],random_seed=42)
```
